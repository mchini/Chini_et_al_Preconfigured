```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()

# where to save the plots
folder4plot = 'E:/PFCmicro/skeweness_paper/figures_v1/fig4_files/'

# load data 
data = pd.read_excel('load graph_stuff.xlsx here')
```

```python
plt.figure()
sns.violinplot(x='age', y='char_path', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='char_path', color='white',
              size=10, edgecolor='black', linewidth=1,
              data=data)
plt.ylim(0.5, 1.3)
```
