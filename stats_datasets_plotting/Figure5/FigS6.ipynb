# import packages

```python
import numpy as np
from scipy.io import loadmat
from scipy.stats import zscore
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
```

```python
# set seaborn theme
sns.set_theme()

# load modeling data
data = pd.read_excel('load df_NN_syms_w_exp.xlsx here')
# set order for plot
order = [0, 1, 2, 3, 'exp. data']
```

# make tSNE kde plot

```python
# make a big matrix for tSNE
X = data[['gini_f', 'gini_s', 'f_s_corr', 'kurt_f', 'kurt_s', 'skew_f', 'skew_s']]
# remove NaNs
idx2keep = np.isnan(X).sum(axis = 1) == 0
X = X.loc[idx2keep]
# apply TSNE
X_embedded = TSNE().fit_transform(X)
```

```python
plt.figure()
sns.kdeplot(x=X_embedded[:, 0], y=X_embedded[:, 1],
            fill=True, hue=data.skewed_params[idx2keep],
            common_norm=False, palette='deep',
            hue_order=order, thresh=0.1)
```
