```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()

# load data and get remove of animals with less than 19 units
data = pd.read_excel('load df_sua_stats_shape.xlsx here')
data = data.loc[data.num_units > 9]
# only keep the relevant part of the dataframe and reorder it
df = data.iloc[:, 4:]
df = df.iloc[:, np.r_[3, 4, 5, 6, 0, 1, 2]]
```

```python
plt.figure()
sns.pairplot(df, dropna=True)
```
