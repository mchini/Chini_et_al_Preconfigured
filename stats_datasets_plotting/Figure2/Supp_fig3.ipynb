# import packages and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()

# load data
data = pd.read_excel('laod df_gini_age.xlsx here')
# fix labels of brain areas
data['b_area'] = data['b_area'].apply(str)
data['b_area'] = data['b_area'].replace({'1':'OB', '2':'PFC'})
# fix colors
colors = ['#ff9900', '#494d87']
order = ['OB', 'PFC']
```

```python
plt.figure()
sns.stripplot(x='age', y='skew_f', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.ylim(1.5, 4.5)
```

```python
plt.figure()
sns.stripplot(x='age', y='skew_s', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.ylim(0, 12)
```

```python
plt.figure()
sns.stripplot(x='age', y='kurt_f', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.yscale('log')
plt.ylim(3, 50)
```

```python
plt.figure()
sns.stripplot(x='age', y='kurt_s', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.yscale('log')
plt.ylim(1, 1000)
```

```python
plt.figure()
sns.stripplot(x='age', y='gini_f', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.ylim(0.5, 0.8)
```

```python
plt.figure()
sns.stripplot(x='age', y='gini_s', hue='b_area',
              dodge=False, jitter=0.05, data=data, 
              palette=colors, hue_order=order,
              legend=False, size=15)
plt.ylim(0, 0.25)
```
