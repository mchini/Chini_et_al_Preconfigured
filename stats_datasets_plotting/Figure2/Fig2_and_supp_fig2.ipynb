# import packages and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()

# load data and get remove animals with less than 19 units
data = pd.read_excel('load df_sua_stats_shape.xlsx here')
data = data.loc[data.num_units > 19]
# fix labels of brain areas
data['b_area'] = data['b_area'].apply(str)
data['b_area'] = data['b_area'].replace({'1':'OB', '2':'PFC'})
# set colors
colors = ['#ff9900', '#494d87']
order = ['OB', 'PFC']
```

```python
plt.figure()
sns.violinplot(x='age', y='skew_f_rate', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='skew_f_rate', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.ylim(-0.5, 6)
```

```python
plt.figure()
sns.violinplot(x='age', y='skew_sttc', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='skew_sttc', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.ylim(-0.5, 15)
```

```python
plt.figure()
sns.violinplot(x='age', y='kurt_f_rate', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='kurt_f_rate', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.yscale('log')
plt.ylim(1, 100)
```

```python
plt.figure()
sns.violinplot(x='age', y='kurt_sttc', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='kurt_sttc', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.yscale('log')
plt.ylim(1, 1000)
```

```python
plt.figure()
sns.violinplot(x='age', y='gini_f', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='gini_f', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.ylim(0.2, 0.9)
```

```python
plt.figure()
sns.violinplot(x='age', y='gini_s', cut=0,
              data=data, palette='cividis_r',
              inner=None, scale='width')
sns.stripplot(x='age', y='gini_s', color='white',
              size=7, edgecolor='black', linewidth=1,
              data=data)
plt.ylim(0, 0.7)
```
