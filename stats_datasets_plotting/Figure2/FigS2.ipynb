# import packages and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()
# load data
data = pd.read_csv('load pvals_thr.csv here')
data = data.iloc[:, 1:]
# transform pvals with -log10
data = -np.log10(data)
vmin = -np.log10(0.01) - 2
vmax = -np.log10(0.01) + 2
```

```python
plt.pcolor(np.flipud(data), cmap='RdBu_r',
           vmin=vmin, vmax=vmax)
plt.colorbar()
```
