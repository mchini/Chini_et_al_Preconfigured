# import packages and data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# set seaborn theme
sns.set_theme()

# load data
data = pd.read_excel('load df_gini.xlsx here')
data['age_rand'] = data['age'] + np.random.rand(data.age.shape[0]) - 0.5
```

```python
plt.figure()
g = sns.scatterplot(x='age_rand', y='skew_f_rate', hue='age', data=data,
                    palette='cividis_r', legend=False, edgecolor='black', linewidth=1)
g.collections[0].set_sizes([100])
plt.ylim(-0.5, 6)
```

```python
plt.figure()
g = sns.scatterplot(x='age_rand', y='skew_sttc', hue='age', data=data,
                    palette='cividis_r', legend=False, edgecolor='black', linewidth=1)
g.collections[0].set_sizes([100])
plt.ylim(-0.5, 5)
```
